```python
# Устанавливаем все неоюходимые библиотеки
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from autofeat.autofeat import AutoFeatRegressor
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel('feautre.xlsx') #Подгружаем датасет
```

```python
#В качетсве таргетного значения выберем topological polar surface area
y = data[['TPSA']]
X = data[['XLogP',   'ExactMass',  'MonoisotopicMass',  'Complexity',  'Charge', 'HBondDonorCount',  'HBondAcceptorCount',  'RotatableBondCount',  'HeavyAtomCount', 'FeatureAcceptorCount3D',  'FeatureDonorCount3D',  'FeatureAnionCount3D', 'FeatureCationCount3D',  'FeatureRingCount3D',  'FeatureHydrophobeCount3D', 'ConformerModelRMSD3D',  'EffectiveRotorCount3D',  'ConformerCount3D', 'MaxEStateIndex' ]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=50)
```

```python
model = AutoFeatRegressor(apply_pi_theorem=True, categorical_cols=None,
          feateng_cols=None, feateng_steps=2, featsel_runs=3,
          transformations=('exp', 'log', 'abs', 'sqrt', '^2', '^3', '1/'),)# указываем допустимый список трансформаций которые будет применять алгоритм
model
```

```python
X_train_feature_creation = model.fit_transform(X_train.to_numpy(), y_train.to_numpy().flatten())
```

```python
X_test_feature_creation = model.transform(X_test.to_numpy())
```
